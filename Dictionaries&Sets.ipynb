# Dictionaries & Sets

## Count frequency of elements in a list using dictionary.

```python
def checkFrequency(arr):
    d = {}
    for i in arr:
        if i in d.keys():
            d[i] += 1
        else:
            d[i] = 1
    return d

checkFrequency([1, 2, 2, 3, 4, 1, 2, 5, 3, 1])
```

## Sort dictionary by values.

```python
def sort_by_value(data):
    # Convert dictionary to list of (key, value) pairs
    items = list(data.items())

    # Bubble sort based on value
    for i in range(len(items)):
        for j in range(0, len(items) - i - 1):
            if items[j][1] > items[j + 1][1]:
                # Swap
                items[j], items[j + 1] = items[j + 1], items[j]

    # Convert back to dictionary
    sorted_dict = {}
    for key, value in items:
        sorted_dict[key] = value
    return sorted_dict

data = {
    "apple": 5,
    "banana": 2,
    "cherry": 8,
    "date": 3
}

sort_by_value(data)
```

## Merge two dictionaries.

```python
def mergeDictionaries(d1,d2):
    d1.update(d2)
    return d1

mergeDictionaries({"a": 1, "b": 2, "c": 3},{"d": 4, "e": 5})
```

## Invert a dictionary.

```python
def invDict(d):
    e = {}
    for i,j in d.items():
        e[j] = i
    return e

invDict({"a": 1, "b": 2, "c": 3})
```

## Find key with maximum value.

```python
def max_value_key(data):
    max_key = None
    max_value = float('-inf')

    for key in data:
        if data[key] > max_value:
            max_value = data[key]
            max_key = key
    return max_key

max_value_key({"apple": 5,"banana": 2,"cherry": 8,"date": 3})
```

## Remove a key from dictionary safely.

```python
def remKey(d, k):
    if d.get(k):
        del d[k]
    return d

remKey({"apple": 5,"banana": 2,"cherry": 8,"date": 3} , "cherry")
```

## Find common keys in two dictionaries.

```python
def checkCommonKey(d1, d2):
    ck = []
    for i in d1.keys():
        if i in d2.keys():
            ck.append(i)
    return ck

checkCommonKey({"name": "Alice", "age": 25, "city": "New York", "email": "alice@example.com"}, 
               {"age": 30, "city": "Chicago", "country": "USA", "phone": "1234567890"})
```

## Create dictionary from two lists.

```python
def createDictFromList(l1,l2):
    d = {}
    for i, j in zip(l1,l2):
        d[i] = j
    return d

createDictFromList(["name", "age", "city"],  ["Alice", 25, "New York"])
```

## Count word frequency in a paragraph.

```python
def countWordFrequency(p):
    d = {}
    for i in p.split(" "):
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    return d

countWordFrequency("Python is great. Python is easy to learn. Learning Python is fun.")
```

## Remove duplicates using set

```python
def remDuplicates(l):
    return list(set(l))

remDuplicates([1, 2, 2, 3, 4, 4, 5, 1, 6])
```
