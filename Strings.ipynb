# Strings

## Reverse a string.

```python
def revstr(s):
    return s[::-1]
revstr("Sky")
```

```python
def revString(s):
    rev_str = ""
    for i in s:
        rev_str = i+rev_str
    return rev_str

revString("Morgan")
```

## Check if a string is palindrome.

```python
def checkPalindrome(x):
    if str(x)==str(x)[::-1]:
        return "Given number is Palindrome"
    else:
        return "Given number is not Palindrome"

checkPalindrome(2345432)
```

## Count vowels and consonants in a string.

```python
def countVowelConsonents(s):
    vowels, consonents = 0, 0
    for i in s:
        if i in ["a", "A", "e","E", 'i','I','o','O','u','U']:
            vowels += 1
        else:
            consonents += 1
    print(f"No. of vowels are {vowels}. \nNo. Consonents are {consonents}.")
countVowelConsonents("CoderhUb")
```

## Find first non-repeating character.

```python
def first_non_repeating(s):
    char_count = {}

    # Count frequency
    for char in s:
        char_count[char] = char_count.get(char, 0) + 1

    # Find first character with count 1
    for char in s:
        if char_count[char] == 1:
            return char

    return None

print(first_non_repeating("swiss"))  # w
```

##  Find duplicate characters in a string.

```python
def findDuplicateStr(s):
    chars = list(s)
    seen = []
    duplicates = []

    for i in chars:
        if i in seen:
            if i not in duplicates:
                duplicates.append(i)
        else:
            seen.append(i)

    return f"Duplicate characters in given string are {duplicates}"


print(findDuplicateStr("Agrahara"))
```

##  Check if two strings are anagrams.

```python
def checkAnagrams(s1, s2):
    if sorted(list(s1.lower())) == sorted(list(s2.lower())):
        return "Given strings are anagram"
    else:
        return "Given strings are not anagram"

checkAnagrams("ram", "aRm")
```

## Remove duplicates from a string.

```python
def removeDuplicateStr(s):
    chars = list(s.lower())
    seen = []
    duplicates = []

    for i in chars:
        if i in seen:
            if i not in duplicates:
                duplicates.append(i)
        else:
            seen.append(i)

    return f"String after removing duplicates : {str(''.join(seen)).capitalize()}"


print(removeDuplicateStr("Agrahara"))
```

## Find frequency of each characters in a string.

```python
def countChars(s):
    charCount = {}
    for i in s:
        if i in charCount.keys():
            charCount[i] += 1
        else:
            charCount[i] = 1
    return charCount

countChars("Sharma")
```

## Check if one string is rotation of another.

```python
def is_rotation(s1, s2):
    # Length must be same and not empty
    if len(s1) != len(s2) or len(s1) == 0:
        return False
    
    return s2 in (s1 + s1)

# is_rotation("hello", "lohe")
is_rotation("python", "thonpy")
```

## Convert string to title case without using built-in methods.

```python
def titleCase(s):
    s = s[0].upper() + s[1:]
    return s

titleCase("deadshot")
```

```python
def titleCase(s):
    result = ""
    new_word = True

    for char in s:
        if char == " ":
            result += char
            new_word = True
        elif new_word:
            # Convert lowercase letter to uppercase manually
            if 'a' <= char <= 'z':
                result += chr(ord(char) - 32)
            else:
                result += char
            new_word = False
        else:
            result += char

    return result


titleCase("deadshot killer")
# Deadshot Killer
```
