## Find largest and smallest element in a list.

```python
def findLargestSmallestElement(arr):
    d = {}

    # Store string and its length
    for i in arr:
        if i not in d:
            d[i] = len(i)

    # Initialize using first element
    largest = smallest = arr[0]

    for key in d:
        if d[key] > d[largest]:
            largest = key
        if d[key] < d[smallest]:
            smallest = key

    return largest, smallest


words = ["apple", "bat", "elephant", "dog"]
findLargestSmallestElement(words)
```

```python
def find_longest_smallest(words):
    if not words:
        return None, None

    longest = smallest = words[0]

    for word in words:
        if len(word) > len(longest):
            longest = word
        if len(word) < len(smallest):
            smallest = word

    return longest, smallest


words = ["apple", "bat", "elephant", "dog"]
longest, smallest = find_longest_smallest(words)

print("Longest:", longest)
print("Smallest:", smallest)
```

```python
words = ["apple", "bat", "elephant", "dog"]

longest = max(words, key=len)
smallest = min(words, key=len)

print("Longest:", longest)
print("Smallest:", smallest)
```

## Find second largest element.

```python

def getSecondlargestword(words):
    for i in range(len(words)):
        for j in range(0, len(words) - i - 1):
            if len(words[j]) < len(words[j + 1]):
                words[j], words[j + 1] = words[j + 1], words[j]
    return words[1]
getSecondlargestword(["apple", "elephant", "bat", "dog"])
```

```python
def getSecondLargest(arr):
    for i in range(len(arr)):
        for j in range(i, len(arr)):
            if arr[i] < arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
    return arr[1]
getSecondLargest([4, 7, 1, 9, 3])
```

```python
def getSecLargest(arr):
    arr.sort()
    return arr[len(arr)-2]
getSecLargest([4, 7, 1, 9, 3])
```

## Rotate a list by k positions.

```python
def posKrotation(k, arr):
    if k == 0:
        return arr
    else:
        return arr[k:] + arr[:k]

posKrotation(3, [5, 7, 12, 15, 18, 25, 40])
```

## Move all zeros to end.

```python
def moveZeroToEnd(arr):
    zeros = []
    non_zeros = []

    for i in arr:
        if i == 0:
            zeros.append(i)
        else:
            non_zeros.append(i)
    non_zeros.extend(zeros)
    return non_zeros

moveZeroToEnd([0, 3, 0, 5, 0, 2, 1])
```

## Find missing number in a list from 1 to n.

```python
def missingValue(arr):
    totalSum = 0
    for i in range(1,len(arr)+2):
        totalSum += i
    return totalSum - sum(arr)

missingValue([1, 2, 3, 4, 5, 7, 8, 9, 10])
```

## Find duplicate elements in a list.

```python
def showDuplicates(arr):
    arr2, arr3 = [], []
    for i in arr:
        if i in arr2:
            arr3.append(i)
        else:
            arr2.append(i)
    return arr3

showDuplicates([43, 72, 11, 19, 43,72, 98, 19, 90, 76, 23, 23, 12,44])
```

## Merge two sorted lists.

```python
def mergeSortedList(arr1, arr2):
    arr1.extend(arr2)

    for i in range(len(arr1)):
        for j in range(i,len(arr1)):
            if arr1[i] > arr1[j]:
                arr1[i], arr1[j] = arr1[j], arr1[i]
    return arr1

mergeSortedList([1, 3, 5, 7], [2, 4, 6, 8])
```
