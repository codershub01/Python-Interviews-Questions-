## Swap two variables without using a third variable 

```python

def swapValues(num1, num2):
    print("num1 before swapping :", num1)
    print("num2 before swapping :", num2)
    print("-------------------------------")
    ## swapping without using any third variable
    num1, num2 = num2, num1
    print("num1 after swapping :", num1)
    print("num2 after swapping :", num2)

swapValues(23, 54)
```

##  Check if a number is even or odd.

```python
def checkOddEven(num):
    if num%2==0:
        print("Given number is even")
    else:
        print("Given num is odd")

checkOddEven(87)
```

## Check if a given number is prime or not

```python
def is_prime(n):
    # Prime numbers are greater than 1
    if n <= 1:
        return False
    
    # Check divisibility from 2 to square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    
    return True

num = int(input("Enter a number: "))

if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

## Find a factorial of a number

```python
def findFactorial(num):
    res = num
    if num<1:
        return "Invalid number"
    elif num==1 :
        return f"Factorial of a given number is {res}"
    for i in range(1, num):
        res *=  i
    print(f"Factorial of a given number is {res}")
findFactorial(6)
```

## Generate Fibonacci series up to n

```python
def fibonacci(n):
    a, b = 0, 1
    count = 0

    if n <= 0:
        print("Please enter a positive integer.")
    elif n == 1:
        print("Fibonacci sequence up to", n, "term:")
        print(a)
    else:
        print("Fibonacci sequence:")
        while count < n:
            print(a, end=" ")
            a, b = b, a + b
            count += 1

fibonacci(5)
```

## Reverse a number

```python
def reverseNum(num):
    return int(str(num)[::-1])
reverseNum(76543)
```

```python
def revNum(num):
    revnum = 0
    
    while num > 0:
        digit = num % 10
        revnum = revnum * 10 + digit
        num //= 10
    
    print(revnum)  # 54321

revNum(886022)
```

##  Check if a number is palindrome.

```python
def checkPalindrome(x):
    if str(x)==str(x)[::-1]:
        return "Given number is Palindrome"
    else:
        return "Given number is not Palindrome"

checkPalindrome(2345432)
```

## Count digits in a number

```python
def countDigits(num):
    return f"No. of digits in given number is {len(str(num))}"


countDigits(87654)
```

## Find sum of  digit of number

```python
def sumOfDigits(num):
    s = 0
    for i in str(num):
        s+=int(i)
    return s
sumOfDigits(12345)
```
